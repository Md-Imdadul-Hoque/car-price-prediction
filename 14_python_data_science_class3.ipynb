**Handling the missing value**: One of the most controversial topic!

**<font color='red'>How to deal with the misssing value?</font>**

Really depends on the nature of the dataset. We can use one of the following approaches:

1. Use the deault fillna metheod -> fill the missing value by the average of a column
2. fill the missing value by the mode or median of the dataset
3. Extrapolate
4. Drop the column altogether
5. Get back to the source and try to find out why the value is missing.
6. Sometimes, the missing value itself can be a feature !

We can write a function to fill the missing value with the most popular methods:

1. For Numericals, fill with the mean
2. For categorical, fill with the mode (the most frequent one)

**<font color='red'>Caution: Do not do it for all the cases, things may change case by case ! </font>**

```python
#function to fill the missing value

def fillmissing_numeric(table, col_name):
    table[col_name] = table[col_name].astype('float64')
    table[col_name].fillna(table[col_name].mean(), inplace=True)
    
def fillmissing_object(table, col_name):
    table[col_name] = table[col_name].astype('object')
    table[col_name].fillna(table[col_name].value_counts().idxmax(), inplace=True)
    # or you can use this instead:
    # table[col_name].fillna(df['num-of-doors'].mode()[0], inplace=True)
```

```python
# fill the numeric values
fillmissing_numeric(df, 'normalized-losses')
fillmissing_numeric(df, 'bore')
fillmissing_numeric(df, 'stroke')
fillmissing_numeric(df, 'horsepower')
fillmissing_numeric(df, 'peak-rpm')
fillmissing_numeric(df, 'stroke')

#fill the categorical values
fillmissing_object(df, 'num-of-doors')
```
