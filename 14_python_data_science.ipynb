## Creating the datetime index

```python
#generating all days
pd.date_range(start='01/01/2020', end='03/01/2020')
```

```python
#generating business days
dates = pd.date_range(start='01/01/2020', end='03/01/2020', freq='B')
dates
```

```python
#using the generated date as index
import numpy as np
random_numbers = np.random.randint(1, 20, len(dates))
random_num_series = pd.Series(random_numbers, index=dates)
pd.DataFrame(random_num_series)
```

### We are in Bangladesh, so we want deshi business day. That means, Friday and Saturday should be the off days

```python
weekmask_Bd = 'Sun Mon Tue Wed Thu'
business_bd = pd.offsets.CustomBusinessDay(weekmask=weekmask_Bd)
```

```python
dates = pd.date_range(start='01/01/2020', end='07/01/2020', freq=business_bd)
```

```python
dates
```

## Addressing the issue of various date formats

```python
different_dates = ['06-21-2020', 'June 21, 2020', '21 June 2020', '06/21/2020', '21st June 2020']
pd.to_datetime(different_dates)
```

```python
pd.to_datetime('21-06-2020')
```

great ! But the problem appears in the ambiguous cases

```python
different_dates = ['January 11, 2020', '11-01-2020']
pd.to_datetime(different_dates)
```

```python
pd.to_datetime(different_dates, dayfirst=True) #dayfirst can clarify the ambiguity
```
